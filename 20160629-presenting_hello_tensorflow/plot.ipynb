```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-0.4, 1.2, 200)
y = x**2
```

```python
x_ = 0.8 + np.linspace(-1, 1, 100)
y_ = 0.64 + 1.6 * np.linspace(-1, 1, 100)
```

```python
plt.plot(x, y, label='loss')
plt.plot([0.8], [0.64], 'bo')
plt.plot(x_, y_, color='r')
plt.xlim([-0.4, 1.2])
plt.ylim([-0.2, 1.4])
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(loc='upper left')
ax = plt.gca()
ax.set_xticks([0.0, 0.8])
ax.set_yticks([0.0, 0.64])
plt.xlabel('$y=w$ for this example')
plt.ylabel('loss')
```
